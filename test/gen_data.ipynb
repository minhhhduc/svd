```python
list_data_shape = [
    (3000, 1024),
    (3333, 768),
    (490, 64),
    (200, 128),
    (500, 256),
    (750, 512),
    (1000, 178)
]
```

```python
import numpy as np

matrix = np.random.rand(10, 10) * 1000
fileName = 'matrix.txt'

with open(fileName, 'w') as f:
    f.write(f'{matrix.shape[0]} {matrix.shape[1]}\n')
    np.savetxt(f, matrix, fmt='%.2f')
```

```python
idx = 0

for sample in range(3):
    for shape in list_data_shape:
        rows, cols = shape
        matrix = np.random.rand(rows, cols) * 1000
        fileName = f'matrix_{idx}.txt'
        
        with open(fileName, 'w') as f:
            f.write(f'{rows} {cols}\n')
            np.savetxt(f, matrix, fmt='%.2f')
        idx += 1
```
